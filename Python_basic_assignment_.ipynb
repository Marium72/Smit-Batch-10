```python
# 12.Password Strength Checker:
```

```python
p=input("Enter your password:")
uppercase_count=0
lowercase_count=0
digits_count=0
for char in p:
  if char.isupper():
    uppercase_count=uppercase_count+1
  elif char.islower():
    lowercase_count=lowercase_count+1
  else:
    digits_count=digits_count+1
if uppercase_count >= 2 and lowercase_count <= 1 and digits_count == 3:
  print("Password is strong")
else:
  print("Password is weak")
  print("Uppercase count", uppercase_count)
  print("Lowercase count", lowercase_count)
  print("Digits count", digits_count)
```
